```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("..2/datafile.csv")
#df = df[["id","ups","downs","pickup_no"]]
def score(row):
    return row["ups"]-row["downs"]
df['score'] = df.apply(score, axis=1)
#df.drop(["ups","downs"],axis=1,inplace=True)
df = df.sort_values(['id','pickup_no'])
post_dict={}
for post_id in df['id'].unique().tolist():
    post_dict[post_id] = []
for _, row in df.iterrows():
    post_dict[row['id']].append(row['score'])
for p, v in df.groupby('id')['score'].max().sort_values()[-8:-1].items():
    print(p,v)
```
