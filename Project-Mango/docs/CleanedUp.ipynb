```python
import pandas as pd
import numpy as np

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm

df = pd.read_csv("../datafile_final.csv")
def score(row):
    return row["ups"]-row["downs"]

df['score'] = df.apply(score, axis=1)
```

```python
print("Total unique posts:")
len(df.drop_duplicates(['id']))
```

```python
print("Total posts recorded:")
len(df)
```

```python
print("Total upvotes on all posts:")
df.sort_values('pickup_no', ascending=False).drop_duplicates(subset=['id']).sort_index().reset_index(drop=True)['ups'].sum()
```

```python
print("Total comments on all posts:")
df.sort_values('pickup_no', ascending=False).drop_duplicates(subset=['id']).sort_index().reset_index(drop=True)['num_comments'].sum()
```

```python
print("Average post title length:")
round(df.sort_values('pickup_no', ascending=False).drop_duplicates(subset=['id']).sort_index().reset_index(drop=True)['post_length'].mean(),2)
```
